kalkulator menghitung volume kerucut

```python

r = int(input("masukan jari jari dari bangun kerucut (cm) : "))
t = int(input(" masukan tinggi bangun kerucut (cm) : "))

volume = 1/3 * 22/7 * r**2 * t

print(f"\volume dari bangun kerucut yaitu {volume} cm^3")
```
